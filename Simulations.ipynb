### Center Frequency

```python
cf_low = np.load("./dat/cf_data_low.npy")
```

```python
fg = FOOOFGroup()
fg.add_data(cf_low[0],cf_low[1])
fg.fit()
```

```python
for i in [0,5,10,15,20]:
    fg.get_fooof(i).plot()
```

```python
cf_high = np.load("./dat/cf_data_high.npy")

fg = FOOOFGroup()
fg.add_data(cf_high[0],cf_high[1])
fg.fit()

for i in [0,5,10,15,20]:
    fg.get_fooof(i).plot()
```
